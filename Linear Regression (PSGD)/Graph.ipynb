```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
KDD_cup1 = pd.read_csv("D:\\OneDrive\\Desktop\\KDDcup1.csv")
KDD_cup2 = pd.read_csv("D:\\OneDrive\\Desktop\\KDDcup2.csv")
KDD_cup4 = pd.read_csv("D:\\OneDrive\\Desktop\\KDDcup4.csv")
KDD_cup6 = pd.read_csv("D:\\OneDrive\\Desktop\\KDDcup6.csv")
KDD_cup7 = pd.read_csv("D:\\OneDrive\\Desktop\\KDDcup7.csv")


time = []
time.append(KDD_cup1["Time"].iloc[0])
time.append(KDD_cup2["Time"].iloc[0])
time.append(KDD_cup4["Time"].iloc[0])
time.append(KDD_cup6["Time"].iloc[0])
time.append(KDD_cup7["Time"].iloc[0])

print(time)

plt.figure(figsize=(10,3))
plt.plot(range(250), KDD_cup1["TrainRMSE"][0:250],label="process1")
plt.plot(range(250), KDD_cup2["TrainRMSE"][0:250],label="process2")
plt.plot(range(250), KDD_cup4["TrainRMSE"][0:250],label="process4")
plt.plot(range(248), KDD_cup6["TrainRMSE"][0:248],label="process6")
plt.plot(range(250), KDD_cup7["TrainRMSE"][0:250],label="process7")
plt.xlabel("No of Epochs")
plt.ylabel("RMSE (Root Mean Squared Error)")
plt.title("Epochs vs Train RMSE")
plt.legend()
    
plt.figure(figsize=(10,3))
plt.plot(range(250), KDD_cup1["TestRMSE"][0:250],label="process1")
plt.plot(range(250), KDD_cup2["TestRMSE"][0:250],label="process2")
plt.plot(range(250), KDD_cup4["TestRMSE"][0:250],label="process4")
plt.plot(range(248), KDD_cup6["TestRMSE"][0:248],label="process6")
plt.plot(range(250), KDD_cup7["TestRMSE"][0:250],label="process7")
plt.xlabel("No of Epochs")
plt.ylabel("RMSE (Root Mean Squared Error)")
plt.title("Epochs vs Test RMSE")
plt.legend()

a = [1,2,4,6,7]
plt.figure(figsize=(10,3))
plt.plot(a, time)
plt.xlabel("No of Workers")
plt.ylabel("Time (sec)")
plt.title("Process vs Train/Test time")
```
